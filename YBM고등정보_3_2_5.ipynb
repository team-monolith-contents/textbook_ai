클래스 안에 함수 정의하기<br>
</div>

```python
class 자전거 :
    색상 = '빨강'
    속력 = 10
    
    def __init__(self, 색상, 속력):
        self.색상 = 색상
        self.속력 = 속력
    
    def 달리기(self):
        print(self.색상 + ' 자전거가 ' + str(self.속력) + 'km/h로 달립니다.')
        
    def 가속하기(self, 가속):
        print(str(가속) + ' km/h만큼 가속합니다.')
        self.속력 = self.속력 + 가속
        
자전거1 = 자전거('노랑', 10)
# 자전거2 = 자전거()
# 자전거3 = 자전거()
자전거1.달리기()
자전거1.가속하기(2)
```

이진 탐색으로 동물 찾기<br>
</div>

```python
동물 = input('찾으려는 동물은? ')
동물원 = ['고양이','기린','돼지','매머드','뱀','사자','원숭이','코끼리','토끼','팬더','하마','호랑이']

번호 = -1
left = 0
right = len(동물원) - 1

while left <= right:
    mid = (left + right) // 2           # 중간 인덱스 계산
    if 동물원[mid] == 동물:              # 찾으려는 값과 중간값이 같은 경우
        번호 = mid
        break
    elif 동물원[mid] < 동물 :            # 찾으려는 값이 중간값보다 큰 경우
        left = mid + 1
    else:                               # 찾으려는 값이 중간값보다 작은 경우
        right = mid -1
        
if 번호 >= 0:
    print(동물 + '을(를) 만나려면 '+str(번호)+'구역으로 가세요.')
else:
    print(동물 + '은(는) 동물원에 없습니다. ')
```

    <br>
문자열 압축하기<br>
</div>

```python
압축전 = input('압축하려는 문자열 : ')
압축후 = ''
글자수 = 1
prev = 압축전[0]
for char in 압축전[1:]:
    if char == prev:             # 현자 문자(char)와 이전 문자(prev)가 같은 경우
        글자수 = 글자수 + 1
    else:                        #현재 문자와 이전 문자가 다른 경우
        압축후 = 압축후 + prev + str(글자수)
        글자수 = 1
        prev = char

# 마지막 문자열 처리
압축후 = 압축후 + prev + str(글자수)
print(압축후)
```
